## Loop/ciclos "for"

```python
numeros = [1,2,3,4,5]

for numero in numeros:
    print(numero)
```

## Loop/ciclos "while"

```python
contador = 5

while contador > 0:
    print(contador)
    contador = contador - 1
```

## Pruebas lógicas con "if"

```python
lenguaje = "Python"

if lenguaje == "Python":
    print("El mejor lenguaje de programación del mundo")
```

```python
lenguaje = "C++"

if lenguaje == "Python":
    print("El mejor lenguaje de programación del mundo")
```

```python
lenguaje = "C++"

if lenguaje != "Python":
    print("No es el mejor de los lenguajes de programación del mundo")
```
