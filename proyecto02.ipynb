```python
import webbrowser
import pyautogui
import pyperclip
import time

# --- Variables (Asegúrate de que estas variables estén definidas antes en tu script) ---
# ticker = "Apple Inc. (AAPL)"
# maxima = 195.20
# minima = 164.10
# promedio = 179.65

# --- Configuración del correo ---
destinatario = "andresfelipebarragannino88@gmail.com"
asunto = f"Análisis de acciones últimos 6 meses - {ticker}"

mensaje = f"""
Buenas noches Andrés,

En el presente correo te envío el análisis de las acciones de los últimos 6 meses de {ticker}.

Cotización máxima: USD {maxima}
Cotización mínima: USD {minima}
Valor medio: USD {promedio}

¡Estoy pendiente de cualquier duda o comentario que tengas al respecto!

Saludos,
Andrés Felipe Barragán Nino
"""

# 1. Abrir Gmail
webbrowser.open("https://mail.google.com/mail/u/0/#inbox")

# Espera generosa para carga inicial
time.sleep(8) 

# Pausa global entre comandos de PyAutoGUI
pyautogui.PAUSE = 1.5

# 2. Clic en "Redactar"
pyautogui.click(x=70, y=208)
time.sleep(2) 

# 3. Llenar Destinatario
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter") # Anclamos el destinatario
time.sleep(0.5)

# 4. Llenar Asunto
pyautogui.press("tab")
pyperclip.copy(asunto)
pyautogui.hotkey("ctrl", "v")
time.sleep(0.5)

# 5. Llenar Cuerpo del Mensaje
pyautogui.press("tab")
pyperclip.copy(mensaje)
pyautogui.hotkey("ctrl", "v")
time.sleep(1)

# 6. Enviar el correo
# Usamos la coordenada que proporcionaste para el botón de enviar
pyautogui.click(x=818, y=684)

# Esperamos un momento a que Gmail confirme el envío antes de cerrar
time.sleep(3)

# 7. Cerrar la pestaña del navegador
pyautogui.hotkey("ctrl", "w")

print("Reporte enviado y pestaña cerrada correctamente.")
```
